```python
import requests
import pandas as pd
```

```python
# OpenAlex API endpoint and parameters
url = "https://api.openalex.org/works"
params = {
    "search": "machine learning",
    "per-page": 30
}

headers = {
    "User-Agent": "YourName (your_email@example.com)"
}

response = requests.get(url, params=params, headers=headers)
data = response.json()
```

```python
# Extract relevant information from the response
works = data["results"]

for work in works:
    title = work["display_name"]
    authors = [a["author"]["display_name"] for a in work["authorships"]]
    year = work["publication_year"]
    citations = work["cited_by_count"]
```

```python
# Create dataframe for extracted data
rows = []

for work in works:
    rows.append({
        "title": work["display_name"],
        "authors": ", ".join(a["author"]["display_name"] for a in work["authorships"]),
        "year": work["publication_year"],
        "citations": work["cited_by_count"],
        "concepts": ", ".join(c["display_name"] for c in work["concepts"])
    })

df = pd.DataFrame(rows)
df
```
